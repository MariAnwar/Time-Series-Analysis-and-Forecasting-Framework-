# Downloading the Raw dataset of Trentino Grid

```python
import requests
import os

doi = "10.7910/DVN/QLCABU"
api_url = f"https://dataverse.harvard.edu/api/datasets/:persistentId?persistentId=doi:{doi}"

response = requests.get(api_url).json()

# Create download folder in Google Drive
output_dir = "Your Path"
os.makedirs(output_dir, exist_ok=True)

# Loop through the files
for file in response['data']['latestVersion']['files']:
    file_id = file['dataFile']['id']
    file_name = file['dataFile']['filename']

    print(f"Downloading: {file_name}")
    download_url = f"https://dataverse.harvard.edu/api/access/datafile/{file_id}?format=original"

    file_response = requests.get(download_url)
    with open(os.path.join(output_dir, file_name), 'wb') as f:
        f.write(file_response.content)
```
